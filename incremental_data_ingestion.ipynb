```python
dbutils.widgets.text("src", "")
```

```python
src_value = dbutils.widgets.get("src")
```

```python
df = spark.readStream.format("cloudFiles")\
          .option("cloudFiles.format", "csv")\
          .option("cloudFiles.schemaLocation", f"/Volumes/flight_catalog/bronze/bronzevolume/{src_value}/checkpoint")\
          .option("cloudFiles.schemaEvolutionMode", "rescue")\
          .load(f"/Volumes/flight_catalog/raw/rawvolume/{src_value}/")
```

```python
df.writeStream.format("delta")\
              .outputMode("append")\
              .trigger(once=True)\
              .option("checkpointLocation", f"/Volumes/flight_catalog/bronze/bronzevolume/{src_value}/checkpoint")\
              .option("path", f"/Volumes/flight_catalog/bronze/bronzevolume/{src_value}/data")\
              .start()
```
